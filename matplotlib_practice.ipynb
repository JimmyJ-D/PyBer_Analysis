```python
#Creating a line graph
#x=left to right y=right to left

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#THIS IS THE MATLAB METHOD
# Create the plot
#chain the plot() function to plt. Inside the parentheses of the plot() function,
#add the x_axis and y_axis parameters:
#The purpose of plt.plot() is to create the figure with the current axes.
plt.plot(x_axis, y_axis)
```

```python
#NOW WE ARE GOING TO USE OBJECT-ORIENTED INTERFACE 

#we switch from using MATLAB to the object-oriented interface method,
#we use ax.plot() instead of plt.plot()

#To create a figure, use fig, ax = plt.subplots().
#To define the axes, use ax.plot().
#Inside the parentheses of ax.plot(), add the axes separated by a comma.

```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

#BREAK DOWN OF THE CODE AND WHAT IS HAPPENING
#The plt.subplots() function returns a tuple that contains a figure and axes object(s).
#fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
#The first variable, fig, references the figure.
#The second variable, ax, references the axes of the figure and allows us to annotate the figure.
#The ax.plot(x_axis, y_axis) plots the data from the x and y axes.

#THE IMAGES ARE THE SAME
```

```python
#THIS IS ANOTHER WAY TO WRITE THE CODE ABOUVE

#Where fig references the figure, and ax references the axes of the figure and allows us to annotate the figure.

# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

```

```python
#Creat a Bubble Chart Using the MATLAB
#useful when presenting financial, population, and weather data because you can add a third and 
#fourth factor to convey more information.

#By changing the "dot" into a "bubble," we are adding a third factor: size.
#If there is more than one dataset that uses the same axes, we can change the color of each marker for each dataset, 
#which will add a fourth factor: color.
#modify the scatter chart to make the bubble chart
```

```python
#Next, to change the size of our marker on the graph, we have to use the s, or size parameter, for the marker. T
#he s parameter must be equal to a feature of our data.
#We will make the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Some of the markers look very small. Let's make them bigger so that the chart is more legible and impactful.

#We can adjust the size by multiplying the data in the y-axis by any value. 
#Let's multiply each data point in the y-axis by 3 and see what happens. 
#To do this, we can iterate through the y-axis data and multiply each data point by 3 and add it to a new list, 
#like this:
#After running this cell, the marker size has changed based on the value of the fare in USD, as you can see here:

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
#Then we can use the new y-axis list for the s parameter:
plt.scatter(x_axis, y_axis, s=y_axis_larger)

#the marker size are 3 time larger than the orginal plot
```

```python
#REFACTORING THE CODE FOR axis_larger using python "list comprehension"
#We can refactor the code we used to create a scatter plot above, 
#but instead of using a for loop to create the list y_axis_larger for the size parameter of the marker, 
#we can use Python's list comprehension technique inside the plt.scatter() function. 
#You can use list comprehension to replace many for and while loops. 
#List comprehension is faster because it is optimized for Python to spot a predictable pattern during looping.

#the format for "list comprehension"
#new_list = [expression for item in list if conditional]

#not total sure I understand this 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```
